## 1. Enunciado del Ejercicio

Imagina que trabajas como analista de datos en una empresa que gestiona datos de ventas
de diferentes tiendas. Se te proporcionan dos conjuntos de datos: ventas.csv y
empleados.csv. El archivo ventas.csv contiene las ventas diarias de diferentes
productos y empleados.csv contiene información sobre los empleados que trabajan en las
tiendas.

Datos de ventas.csv
- id_venta: ID único para cada venta.
- fecha: Fecha en la que se realizó la venta.
- id_empleado: ID del empleado que realizó la venta.
- tienda: Nombre de la tienda donde se realizó la venta.
- producto: Producto que se vendió.
- cantidad: Cantidad de productos vendidos.

Datos de empleados.csv
- id_empleado: ID único para cada empleado.
- nombre: Nombre del empleado.
- tienda: Tienda donde trabaja el empleado.

Tu tarea es realizar las siguientes operaciones:
1. Merge: Combina los dos conjuntos de datos usando el id_empleado como clave
para obtener un nuevo DataFrame que contenga toda la información de las ventas
junto con los nombres de los empleados.
2. Concat: Supón que tienes otro conjunto de datos ventas2.csv que también
incluye ventas. Concatena este nuevo DataFrame con el original ventas.csv para
tener un único DataFrame con todas las ventas.
3. Join: Utiliza la operación de join para combinar los datos de ventas.csv con
empleados.csv, asumiendo que ventas.csv se ha convertido en un DataFrame
llamado df_ventas y empleados.csv en df_empleados. Usa id_empleado
como clave pero esta vez realiza un join izquierdo.
4. GroupBy: Agrupa el DataFrame combinado por tienda y calcula el total de ventas
(suma de cantidad) por tienda.



```python
import pandas as pd
```

```python
ventas1_df = pd.read_csv('/content/drive/MyDrive/datasets/ventas.csv')
empleados_df = pd.read_csv('/content/drive/MyDrive/datasets/empleads.csv')
```

```python
# Merge
ventas_empleados_df = pd.merge(ventas1_df, empleados_df, on='id_empleado')
ventas_empleados_df.head()
```

```python
# Concat
ventas2_df = pd.read_csv('/content/drive/MyDrive/datasets/ventas2.csv')
ventas_df = pd.concat([ventas1_df, ventas2_df])
ventas_df.head()
```

```python
# Join
ventas1_df.join(empleados_df, on='id_empleado', how='left', lsuffix='_x', rsuffix='_y')
```
