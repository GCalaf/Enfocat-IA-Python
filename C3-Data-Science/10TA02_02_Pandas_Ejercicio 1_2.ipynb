#Ejercicio: Análisis de Datos de Empleados

**Enunciado**

Eres el analista de recursos humanos de una empresa que ha registrado la información de
sus empleados en un archivo CSV. La empresa te ha proporcionado un archivo CSV con la
siguiente información:
- employee_id: El identificador del empleado.
- name: El nombre del empleado.
- department: El departamento en el que trabaja el empleado.
- position: El cargo del empleado.
- salary: El salario mensual del empleado.
- date_of_hire: La fecha de contratación del empleado.

Tu tarea es analizar estos datos para responder las siguientes preguntas:
1. ¿Cuál es el salario promedio por departamento?
2. ¿Cuál es el número total de empleados por departamento?
3. ¿Cuáles son los 5 empleados con los salarios más altos?
4. ¿Cuál es el salario promedio de los empleados contratados en los últimos 5 años?
5. ¿Cuáles son los 3 departamentos con los salarios promedios más bajos?



```python
import pandas as pd
```

```python
empleados = pd.read_csv("employees_data.csv", thousands='.')
salario_medio_departamento = empleados.groupby(by='department')['salary'].mean()
salario_medio_departamento
```

```python
empleados.groupby(by='department')['employee_id'].count()
```

```python
empleados.sort_values('salary', ascending=False).head()
```

```python
empleados['date_of_hire'] = pd.to_datetime(empleados['date_of_hire'])
empleados_ultimos_5_anios = empleados[empleados['date_of_hire'] >= pd.Timestamp.now() - pd.DateOffset(years=5)]
empleados_ultimos_5_anios['salary'].mean()
```
