### Ejercicio 2

```python
data = [
 ('John', [45, 55, 60, 70]),
 ('Emily', [88, 76, 95, 80]),
 ('Sophia', [60, 65, 45, 70]),
 ('Michael', [80, 90, 92, 85])
]
```

#### # Usar map para calcular el promedio de calificaciones mayores o iguales a 50 para cada persona.


```python
from statistics import mean
promedio = list(map(lambda x: (x[0], mean(list(filter(lambda y: y >= 50, x[1])))), data))
print(promedio)
```

#### # Convertir a lista y ordenar por el promedio de calificaciones


```python
promedio.sort(key=lambda x: x[1], reverse=True)
print(promedio)
```

#### # Filtrar las personas con el promedio m치s alto


```python
avg = mean(list(map(lambda x: x[1], promedio)))
promedio_max = list(filter(lambda x: x[1] > avg, promedio))
print(promedio_max)
```

#### # Extraer solo los nombres de las personas con el promedio m치s alto


```python
nombres = list(map(lambda x: x[0], promedio_max))
print(nombres)
```

### Completo

```python
from functools import reduce

data = [
 ('John', [45, 55, 60, 70]),
 ('Emily', [88, 76, 95, 80]),
 ('Sophia', [60, 65, 45, 70]),
 ('Michael', [80, 90, 92, 85])
]

promedio = list(map(lambda x: (x[0], reduce(lambda a, b: a + b, x[1]) / len(x[1])), data))
promedio.sort(key=lambda x: x[1], reverse=True)
avg = reduce(lambda a, b: a + b, list(map(lambda x: x[1], promedio))) / len(promedio)
promedio_max = list(filter(lambda x: x[1] > avg, promedio))
nombres = list(map(lambda x: x[0], promedio_max))
print('Los nombres de las personas con el promedio m치s alto son:', nombres)
```
