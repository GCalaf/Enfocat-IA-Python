### Nivel básico

Enunciado:
Escribe una función llamada suma_simple que reciba dos números y devuelva su suma. Asegúrate de que la función maneje correctamente números negativos.

```python
def suma_simple(a:int, b:int) -> int:
  return a + b
suma_simple(1, 2)
```

```python
def suma_simple(a:float, b:float) -> float:
  if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
    raise TypeError("Los argumentos deben ser números")
  return a + b
suma_simple(1.0, 2.0)
```

```python
from typing import Union
def suma_simple(a:Union[int, float], b:Union[int, float]) -> Union[int, float]:
  return a + b
suma_simple(1.0, 2.0)
```

### Nivel Intermedio
Enunciado:
Crea una función llamada filtrar_pares que reciba una lista de números y devuelva una nueva lista solo con los números pares.

```python
def filtrar_pares(lista:list) -> list:
  pares=[]
  for x in lista:
    if x % 2 == 0:
      pares.append(x)
  return pares
filtrar_pares([1,2,3,4,5])
```

```python
def filtrar_pares(lista:list) -> list:
  return [x for x in lista if x % 2 == 0]
filtrar_pares([1,2,3,4,5])
```

```python
def filtrar_pares(lista:list) -> list:
  if not all(isinstance(x, (int, float)) for x in lista):
    raise TypeError("Los argumentos deben ser números")
  return [x for x in lista if x % 2 == 0]
filtrar_pares([1,2,3,4,5])
```

```python
from typing import Union
def filtrar_pares(lista:list) -> list:
  return list(filter(lambda x: x % 2 == 0, lista))
filtrar_pares([1,2,3,4,5])
```

### Nivel Avanzado
Enunciado:
Crea una función llamada factorial que calcule el factorial de un número usando recursividad. El factorial de n se define como n * (n-1) * (n-2) ... * 1.

```python
def factorial(n:int) -> int:
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
factorial(3)
```

```python
def factorial(n:int) -> int:
  if n < 0:
    raise ValueError("El argumento debe ser un número positivo")
  return 1 if n == 0 else n * factorial(n-1)
factorial(3)
```
