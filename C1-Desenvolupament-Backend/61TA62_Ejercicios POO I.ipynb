### Ejercicio 1
Vamos a crear una clase llamada Persona. Sus atributos son: nombre, edad
y DNI. Construye los siguientes métodos para la clase:
- Un constructor, donde los datos pueden estar vacíos.
- Los setters y getters para cada uno de los atributos. Hay que validar
las entradas de datos.
- mostrar(): Muestra los datos de la persona.
- esMayorDeEdad(): Devuelve un valor lógico indicando si es mayor de
edad.

```python
class Persona():
    def __init__(self, nombre:str = "", edad:int = 0, dni:str = ""):
        self._nombre = nombre
        self._edad = edad
        self._dni = dni

    @property
    def _nombre(self):
        return self.nombre
    @property
    def _dni(self):
        return self.dni
    @property
    def _edad(self):
        return self.edad

    @_nombre.setter
    def _nombre(self, nombre:str):
        if not isinstance(nombre, str):
            raise TypeError("El nombre debe ser un string")
        self.nombre = nombre
    @_dni.setter
    def _dni(self, dni:str):
        if not isinstance(dni, str):
            raise TypeError("El DNI debe ser un string")
        self.dni = dni
    @_edad.setter
    def _edad(self, edad:int):
        if not isinstance(edad, int) or edad < 0:
            raise TypeError("La edad debe ser un entero positivo")
        self.edad = edad

    def mostrar(self):
        print(f"Nombre: {self.nombre}\nEdad: {self.edad}\nDNI: {self.dni}")

    def esMayorDeEdad(self) -> bool:
        return self._edad >= 18

juan = Persona("Juan", 30, "12345678F")
juan.mostrar()
juan.esMayorDeEdad()
```

### Ejercicio 2
Crea una clase llamada Cuenta que tendrá los siguientes atributos: titular
(que es una persona) y cantidad (puede tener decimales). El titular será
obligatorio y la cantidad es opcional. Construye los siguientes métodos
para la clase:
- Un constructor, donde los datos pueden estar vacíos.
- Los setters y getters para cada uno de los atributos. El atributo no se
puede modificar directamente, sólo ingresando o retirando diner
- mostrar(): Muestra los datos de la cuenta.
- ingresar(cantidad): se ingresa una cantidad a la cuenta, si la
cantidad introducida es negativa, no se hará nada.
- retirar(cantidad): se retira una cantidad a la cuenta. La cuenta
puede estar en números rojos.

```python
class Cuenta():
    def __init__(self, titular:Persona, cantidad:float = 0):
        self._titular = titular
        self._cantidad = cantidad

    @property
    def _titular(self):
        return self.titular
    @property
    def _cantidad(self):
        return self.cantidad

    @_titular.setter
    def _titular(self, titular:Persona):
        if not isinstance(titular, Persona):
            raise TypeError("El titular debe ser una persona")
        self.titular = titular
    @_cantidad.setter
    def _cantidad(self, cantidad:float):
        if not isinstance(cantidad, (float, int)):
            raise TypeError("La cantidad debe ser un numero")
        self.cantidad = cantidad

    def mostrar(self):
        print(f"Titular: {self.titular.nombre}\nCantidad: {self.cantidad}")

    def ingresar(self, cantidad:float):
        if cantidad < 0:
            return
        self.cantidad += cantidad

    def retirar(self, cantidad:float):
        self.cantidad -= cantidad

cuenta_juan = Cuenta(juan, 1000)
cuenta_juan.mostrar()
cuenta_juan.ingresar(1000)
cuenta_juan.mostrar()
cuenta_juan.ingresar(-1000)
cuenta_juan.mostrar()
cuenta_juan.retirar(500)
cuenta_juan.mostrar()
```

### Ejercicio 3
Vamos a definir ahora una “Cuenta Joven”, para ello vamos a crear u
nueva clase CuantaJoven que deriva de la anterior. Cuando se crea esta
nueva clase, además del titular y la cantidad se debe guardar una
bonificación que estará expresada en tanto por ciento.Construye l
siguientes métodos para la clase:
- Un constructor.
- Los setters y getters para el nuevo atributo.
- En esta ocasión los titulares de este tipo de cuenta tienen que ser
mayor de edad., por lo tanto hay que crear un método
esTitularValido() que devuelve verdadero si el titular es mayor de
edad pero menor de 25 años y falso en caso contrario.
- Además la retirada de dinero sólo se podrá hacer si el titular es
válido.
- El método mostrar() debe devolver el mensaje de “Cuenta Joven” y
la bonificación de la cuent

Piensa los métodos heredados de la clase madre que hay que reescribir.

```python
class CuentaJoven(Cuenta):
    def __init__(self, titular:Persona, cantidad:float = 0, bonificacion:float = 0):
        super().__init__(titular, cantidad)
        self._bonificacion = bonificacion

    @property
    def _bonificacion(self):
        return self.bonificacion

    @_bonificacion.setter
    def _bonificacion(self, bonificacion:float):
        if not isinstance(bonificacion, (float, int)):
            raise TypeError("La bonificacion debe ser un número")
        self.bonificacion = bonificacion

    def esTitularValido(self) -> bool:
        return self.titular.esMayorDeEdad() and self.titular.edad < 25

    def retirar(self, cantidad:float):
        if not self.esTitularValido():
            return print("El titular no es válido")
        super().retirar(cantidad)

    def mostrar(self):
        print(f"Cuenta Joven\nTitular: {self.titular.nombre}\nCantidad: {self.cantidad}\nBonificacion: {self.bonificacion}")


juan_joven = CuentaJoven(juan, 1000, 10)
juan_joven.mostrar()
juan_joven.retirar(500)
juan_joven.mostrar()
```
